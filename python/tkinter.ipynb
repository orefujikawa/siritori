```python
import re
import random
# string=input(prin)
def siri():
    global num
    string=num
    textfile=['1.txt','3.txt','4.txt','5.txt']
    last=string[-1]
    pattern='^'+last+'.'+'*'# pattern='^い*'# print(pattern)
    choicefile=random.choice(textfile)
    file=open(choicefile,'r',encoding='UTF-8')
    Matching=re.compile(pattern)
    deta=file.readlines()
    # print(deta)['あい（愛\n', 'いあつ（威圧\n', 'うい（憂い\n', 'えあこん（エアコン\n', 'おあしす（オアシス\n', 'かあさん（母さん\n']
    for i in deta:
            logic=re.match(Matching,i)# print(i)
            if logic==None:
                continue
            co=logic.group()
            # print(co)
    file.close()
    return co

```
